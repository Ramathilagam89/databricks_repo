```python
datastruct="id int,fname string,lname string, age int,prof string"
df1=spark.read.schema(datastruct).csv("/Volumes/read_write_catalog/rd_wt_schema/rd_wt_volume/rd_wr_dir/custs")
print(df1.printSchema())
display(df1.limit(3))
```

```python
from pyspark.sql.types import StructType,StructField,IntegerType,StringType
column_schema=StructType([StructField("id",IntegerType(),False),StructField("fname",StringType()),StructField("lname",StringType()),StructField("age",IntegerType()),StructField("prof",StringType())])
df1=spark.read.schema(column_schema).csv("/Volumes/read_write_catalog/rd_wt_schema/rd_wt_volume/rd_wr_dir/custs")
df1.printSchema()
display(df1.limit(3))
```

```python
df1.write.csv("/Volumes/read_write_catalog/rd_wt_schema/rd_wt_volume/rd_wr_dir/custs_write.csv",header=True,sep="~",mode="Overwrite")
display(df1.limit(3))   
```

```python
df1.write.json("/Volumes/read_write_catalog/rd_wt_schema/rd_wt_volume/rd_wr_dir/custs_write_json",mode="append")
display(df1.limit(3))
```

```python
df1.write.orc("/Volumes/read_write_catalog/rd_wt_schema/rd_wt_volume/rd_wr_dir/custs_write_orc",mode="append")
df2=spark.read.orc("/Volumes/read_write_catalog/rd_wt_schema/rd_wt_volume/rd_wr_dir/custs_write_orc")
display(df2.limit(3))
```

```python
df1.write.saveAsTable("default.customertbl",mode='overwrite')#default delta format
spark.read.table("default.customertbl").show(2)
```
